## Grafik Düzenleme

### Legend

```python
x = np.arange(0,10)
```

```python
plt.figure(figsize=(10,6), dpi=120)
plt.plot(x, 2*x + x, label = "2x + x")
plt.plot(x, x**2-x, label = "x^2 - x")
plt.legend();
```

* plt.legend(loc= 0) : Matplotlib ideal noktayı seçer
* plt.legend(loc= 1) : Sağ üst
* plt.legend(loc= 2) : Sol üst
* plt.legend(loc= 3) : sol alt
* plt.legend(loc= 4) : Sağ alt

```python
fig = plt.figure(figsize=(10,6), dpi=120)

ax = fig.add_axes([0,0,1,1])

ax.plot(x, 2*x + x, label = "2x + x")
ax.plot(x, x**2-x, label = "x^2 - x")
ax.legend();

# fig değişkeninin içine yukarıdaki grafiği koyduk.
# Artık her seferinde takrar tekrar grafiği yazırmak gerekmiyor
```

Subplotların arasındaki boşlukları isteğe göre düzenlemek için: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1][0].plot(x,x**2)
ax[1,1].plot(x,x**0.5)

fig.subplots_adjust(
    left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.6,
    hspace=0.6,
    )
# plt.tight_layout() ---> Verimsiz olduğu durumda kullanılabilir.
```

## Eksen Düzenlemek

### Logaritmik Eksen

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x,np.exp(x))
ax[0].set_title("Normal Grafik")

ax[1].plot(x,np.exp(x))
ax[1].set_yscale("log")
ax[1].set_title("Logaritmik eksen (y)")

plt.tight_layout();
```

### İki Eksenli Grafik

```python
fig, ax1 = plt.subplots()

ax1.plot(x, 4*x, lw=2, color="blue", marker="o")
ax1.set_ylabel(r"Çevre $(4X)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**2, lw=2, color="red", marker="s")
ax2.set_ylabel(r"Alan $(X^2)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")

```

### Negatif Sayılar Eksene Eklemek

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

x = np.linspace(-4, 3, 100)
y = 1.7*x**3 + 3.7*x**2 - 2.36*x + 7.62
ax.plot(x, y);
```

```python
fig, ax = plt.subplots()

plt.axvline(0, c='grey', ls="--", lw=1)
plt.axhline(0, c='grey', ls="--", lw=1)

ax.plot(x, x**2, x, np.log(x))

ax.text(-3, 9.5, r"$y=x^2$", fontsize=12, color="blue")
ax.text(0.3, -3, r"$y=log(x)$", fontsize=12, color="green")
```

## 3D Grafik

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(figsize=(9,9), dpi=200, subplot_kw={"projection": "3d"})
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```python
fig = plt.figure(figsize=(9,9), dpi=200)

ax = fig.add_subplot(1, 1, 1, projection='3d')

p = ax.plot_wireframe(X, Y, Z, rstride=1, cstride=1)
```
