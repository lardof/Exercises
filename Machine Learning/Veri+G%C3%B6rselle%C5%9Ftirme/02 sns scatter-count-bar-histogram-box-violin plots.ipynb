# **SEABORN**

Seaborn hakkında detaylı bilgi için: https://seaborn.pydata.org/

```python
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("community_health_metric.csv")
# Veri seti eğitim için uydurulmuş bir veri setidir.
# Gerçek kişi ve kurumlarla ilgisi bulunmamaktadır.
```

```python
df
```

```python
df.info()
```

## Scatter Plot

```python
# help(sns.scatterplot)
```

```python
plt.figure(dpi=100)
sns.scatterplot(data= df, x="kilo" , y="boy")
```

```python
plt.figure(dpi=100, figsize=(12,8))
sns.scatterplot(data= df, x="kilo" , y="boy", hue="cinsiyet")
```

```python
plt.figure(dpi=100)
sns.scatterplot(data= df, x="kilo" , y="boy", hue="cinsiyet", palette="colorblind")
```

Farklı renklerdeki paletler için aşağıdaki linklere göz atabilirsin

* https://matplotlib.org/stable/gallery/color/colormap_reference.html

* https://seaborn.pydata.org/tutorial/color_palettes.html

```python
plt.figure(dpi=100, figsize=(10,6))
sns.scatterplot(data= df, x="kilo" , y="boy", hue="cinsiyet", palette="viridis", size="etnik")
```

```python
plt.figure(dpi=100, figsize=(10,6))
sns.scatterplot(data= df, x="kilo" , y="boy", palette="magma",
 s=200, linewidth=0, alpha=0.5)
```

```python
plt.figure(dpi=100, figsize=(10,6))
sns.scatterplot(data= df, x="kilo" , y="boy", palette="magma", hue="cinsiyet",
 s=200, linewidth=0.5, alpha=0.5)
```

```python
plt.figure(dpi=100, figsize=(10,6))
sns.scatterplot(data= df, x="kilo" , y="boy", palette="magma", hue="cinsiyet",
 s=200, linewidth=0.5, alpha=0.7, style="cinsiyet")
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.countplot(data=df,x="etnik");
```

## Bar Plot

```python
# help(sns.barplot)
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.barplot(data=df,x="dr", y="kilo");
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.barplot(data=df,x="etnik", y="boy", ci="sd",estimator=np.mean)
plt.ylim(140,200);
```

```python
plt.figure(figsize=(12,6), dpi=200)
sns.barplot(data=df,x="etnik", y="boy", ci="sd",estimator=np.mean, hue="cinsiyet")
plt.ylim(140,200)
plt.legend(bbox_to_anchor=(1.,0.6))
```

## Histogram

```python
# help(sns.histplot)
```

```python
# help(sns.displot)
```

```python
sns.histplot(data=df, x="dr")
```

```python
sns.displot(df, x="dr");
```

```python
sns.histplot(data=df, x="dr", bins=6)
```

```python
sns.histplot(data=df, x="dr", kde=True);
```

```python
sns.set(style="dark")
sns.histplot(data=df, x="dr", kde=True);
```

Grafik stilleri ve değişimi için bakabilirsiniz: https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme

```python
plt.figure(figsize=(12,6), dpi=200)
sns.set(style="whitegrid")
sns.histplot(data=df, x="dr", kde=True, edgecolor="purple", lw=1.5, ls="--", color="orange");
```

## Boxplot

```python
# help(sns.boxplot)
```

```python
plt.figure(figsize=(12,6), dpi=120)
sns.boxplot(data=df, x="etnik", y="kilo")
```

## Violin Plot

```python
# help(sns.violinplot)
```

```python
sns.violinplot(x="yas",y='dr',data=df)
# çok fazla veri içeren bir şey girerseniz grafiğiniz patates olur
```

```python
sns.violinplot(x="etnik",y='yas',data=df)
```

```python
sns.violinplot(x="etnik",y='yas',data=df, hue="cinsiyet")
plt.legend(bbox_to_anchor=(1.05,0.6))
```

```python
sns.violinplot(x="etnik",y='yas',data=df, hue="cinsiyet", split=True)
plt.legend(bbox_to_anchor=(1.05,0.6))
```

```python
sns.violinplot(x="etnik",y='yas',data=df, hue="cinsiyet", split=True, inner="quartile")
plt.legend(bbox_to_anchor=(1.05,0.6))
```

```python
sns.violinplot(x="etnik",y='yas',data=df, hue="cinsiyet", split=True, bw = 0.1)
plt.legend(bbox_to_anchor=(1.05,0.6))
```

```python
sns.violinplot(x="etnik",y='yas',data=df, hue="cinsiyet", split=True, bw = 2)
plt.legend(bbox_to_anchor=(1.05,0.6))
```
